```python
#TUGAS 4
import pandas as pd

# Data perusahaan
company_detail_list = [
    {'name': 'Company 1', 'domain': 'Retail', 'country': 'United States'},
    {'name': 'Company 2', 'domain': 'Technology', 'country': 'United Kingdom'},
    {'name': 'Company 3', 'domain': 'Healthcare', 'country': 'United States'}
]

# Data karyawan
employee_detail_list = [
    {'name': 'EMP-0001', 'first_name': 'John', 'last_name': 'Doe', 'full_name': 'John Doe', 'company': 'Company 1', 'nationality': 'Australia'},
    {'name': 'EMP-0002', 'first_name': 'Tom', 'last_name': 'Smith', 'full_name': 'Tom Smith', 'company': 'Company 2', 'nationality': 'United States'},
    {'name': 'EMP-0003', 'first_name': 'Andrew', 'last_name': 'Sebastian', 'full_name': 'Andrew Sebastian', 'company': 'Company 3', 'nationality': 'United States'},
    {'name': 'EMP-0005', 'first_name': 'Ying Han', 'last_name': 'Tan', 'full_name': 'Ying Han Tan', 'company': 'Company 1', 'nationality': 'Australia'},
    {'name': 'EMP-0015', 'first_name': 'Kenneth', 'last_name': 'Ng', 'full_name': 'Kenneth Ng', 'company': 'Company 3', 'nationality': 'United States'},
    {'name': 'EMP-0018', 'first_name': 'Rubby', 'last_name': 'Lee', 'full_name': 'Rubby Lee', 'company': 'Company 2', 'nationality': 'Hong Kong'},
    {'name': 'EMP-0017', 'first_name': 'Robert', 'last_name': 'White', 'full_name': 'Robert White', 'company': 'Company 1', 'nationality': 'United Kingdom'}
]

def group_employees_by_country(employee_detail_list):
    # Menyiapkan hasil berdasarkan negara
    grouped_by_country = {}
    
    for employee in employee_detail_list:
        country = employee['nationality']  # Menggunakan nationality karyawan
        
        # Jika negara belum ada di dictionary, buat entry baru
        if country not in grouped_by_country:
            grouped_by_country[country] = []
        
        # Tambahkan karyawan ke negara yang sesuai
        grouped_by_country[country].append({
            'full_name': employee['full_name'],
            'company': employee['company']
        })
    
    return grouped_by_country

# Mendapatkan karyawan yang dikelompokkan berdasarkan kewarganegaraan
employees_grouped_by_country = group_employees_by_country(employee_detail_list)

# Menampilkan hasil
for country, employees in employees_grouped_by_country.items():
    print(f"Negara: {country}")
    for emp in employees:
        print(f"  Nama: {emp['full_name']}, Perusahaan: {emp['company']}")
```

```python
#TUGAS 5
import pandas as pd

# Data
employee_detail_list = [
    {'name': 'EMP-0001', 'first_name': 'John', 'last_name': 'Doe', 'full_name': 'John Doe', 'company': 'Company 1', 'nationality': 'Australia'},
    {'name': 'EMP-0002', 'first_name': 'Tom', 'last_name': 'Smith', 'full_name': 'Tom Smith', 'company': 'Company 2', 'nationality': 'United States'},
    {'name': 'EMP-0003', 'first_name': 'Andrew', 'last_name': 'Sebastian', 'full_name': 'Andrew Sebastian', 'company': 'Company 3', 'nationality': 'United States'},
    {'name': 'EMP-0005', 'first_name': 'Ying Han', 'last_name': 'Tan', 'full_name': 'Ying Han Tan', 'company': 'Company 1', 'nationality': 'Australia'},
    {'name': 'EMP-0015', 'first_name': 'Kenneth', 'last_name': 'Ng', 'full_name': 'Kenneth Ng', 'company': 'Company 3', 'nationality': 'United States'},
    {'name': 'EMP-0018', 'first_name': 'Rubby', 'last_name': 'Lee', 'full_name': 'Rubby Lee', 'company': 'Company 2', 'nationality': 'Hong Kong'},
    {'name': 'EMP-0017', 'first_name': 'Robert', 'last_name': 'White', 'full_name': 'Robert White', 'company': 'Company 1', 'nationality': 'United Kingdom'}
]

# Fungsi untuk menghitung kewarganegaraan karyawan berdasarkan perusahaan
def count_employee_nationalities_by_company(employee_detail_list):
    result = []
    company_nationalities = {}

    for employee in employee_detail_list:
        company = employee['company']
        nationality = employee['nationality']
        
        # Jika perusahaan belum ada di dictionary, buat entry baru
        if company not in company_nationalities:
            company_nationalities[company] = {}
        
        # Jika kewarganegaraan sudah ada, tambahkan hitungannya
        if nationality in company_nationalities[company]:
            company_nationalities[company][nationality] += 1
        else:
            company_nationalities[company][nationality] = 1

    # Mengonversi dictionary menjadi list untuk hasil akhir
    for company, nationalities in company_nationalities.items():
        result.append({
            'company': company,
            'employee_nationality': nationalities
        })
    
    return result

# Mendapatkan jumlah kewarganegaraan karyawan berdasarkan perusahaan
nationality_counts = count_employee_nationalities_by_company(employee_detail_list)

# Membuat DataFrame untuk menampilkan hasil
nationality_counts_df = pd.DataFrame(nationality_counts)

# Menampilkan DataFrame
display(nationality_counts_df)
```
